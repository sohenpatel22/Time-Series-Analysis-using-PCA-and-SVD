Climate Trends (1900–2012)

X-Axis: Year (1900 to just after 2010)

Y-Axis: Value (ranging from roughly -20 to 30)

Key Trend: The vast majority of the data clusters densely between 0 and 30, showing a slight general upward trend in the later decades, while a couple of distinct outlier lines remain consistently low near -20.
